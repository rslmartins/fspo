# 3 - Categorize equipment failures by their nature/root cause according to parameter readings (temperature, pressure, and others)



## Generally in maintenance engineering, in this case, data scientists try to detect anomalies from a given dataset. In fact, it means trying to find data that don't behaves as a desired pattern, the outliers. Therefore, outlier detection can be done using statistical methods or using a data science approach. For example, as seen below, box-plot is a famous method to detect from outliers using quartile property.

```python
df[df.columns[3:9]].boxplot(figsize=(800*pixel, 600*pixel))
plt.show()
```

## Then, after using a visual approach, it is possible to classify each feature as an outlier by using a threshold of third quartile plus 1.5 interquartile.

```python
def check_outliers(k = 1.5):
    stats_parameters = df[df.columns[3:9]].describe()
    upper_dict = {}
    for column in stats_parameters.columns:
        IQR = stats_parameters[column]['75%']-stats_parameters[column]['25%']
        upper_dict[column] = stats_parameters[column]['75%'] + k*(IQR)

    stats_parameters = stats_parameters.append(pd.Series(data=upper_dict, name ='upper extreme'))
    display(stats_parameters)
    

    columns = df.columns[3:9]
    for column in columns:
        outlier_class = {}
        for i in df_Fail[column].index:
            if df_Fail[column][i] <= stats_parameters[column]['25%']: outlier_class[i] = 'Q1'
            if df_Fail[column][i] > stats_parameters[column]['25%'] and df_Fail[column][i] <= stats_parameters[column]['50%']: outlier_class[i] = 'Q2'
            if df_Fail[column][i] > stats_parameters[column]['50%'] and df_Fail[column][i] <= stats_parameters[column]['75%']: outlier_class[i] = 'Q3'
            if df_Fail[column][i] > stats_parameters[column]['75%'] and df_Fail[column][i] <= stats_parameters[column]['upper extreme']: outlier_class[i] = 'Q4'
            if df_Fail[column][i] > stats_parameters[column]['upper extreme']: outlier_class[i] = 'outlier'
        outlier_series = pd.Series(data=outlier_class, name =column+'_outlier')
        outlier_series.index = outlier_series.index.set_names('Cycle')

        df_Fail[column+'_outlier'] = outlier_series
            
    df_Fail['Nature/Root'] = None
    for i in df_Fail.index:
        root = ''
        if df_Fail['Temperature_outlier'][i] == 'outlier': root = root+'Temperature '
        if df_Fail['Pressure_outlier'][i] == 'outlier': root = root+'Pressure '
        if df_Fail['VibrationX_outlier'][i] == 'outlier': root = root+'VibrationX '
        if df_Fail['VibrationY_outlier'][i] == 'outlier': root = root+'VibrationY '
        if df_Fail['VibrationZ_outlier'][i] == 'outlier': root = root+'VibrationZ '
        if df_Fail['Frequency_outlier'][i] == 'outlier': root = root+'Frequency '
        df_Fail['Nature/Root'][i] = root
        if df_Fail['Nature/Root'][i] == '': df_Fail['Nature/Root'][i] = 'Unkown'

    display(df_Fail['Nature/Root'].value_counts())
```

## From the result below, the majority of the failures has unknown nature/root. Which means that a further investigation is demanded. For example, combining features to check its behavior (e.g. VibrationZ divided by Frequency), transforming features (e.g. squared root of Temperature) and finally trying different outlier detection methods: distance-based, clustering, classification techniques

# 4 – Create a model using the technique you think is most appropriate and measure its performance

```python
counters = df['Fail Counter'].unique()
counter = counters[0]
for counter in counters:
    if counter < counters[-1]:
        len_train = df[df['Fail Counter']==counter & (df['Fail']==False)].shape[0]
        len_test = df[(df['Fail Counter']==counter+1) & (df['Fail']==True)].shape[0]
        print('Fail Counter', counter)
        print('Total length', len_train+len_test)
        print('Train',round(len_train/(len_train+len_test)*100,2))
        print('Test',round(len_test/(len_train+len_test)*100,2))
        #print(len_test/(len_train+len_test)))
        print('\t')
```

## In order to select a good database is prefered to have a significant length and also have a good balance between train/test. From the 9 datasets with a 'health' system and a failured system, Failure Counter of 5 fits better.

```python
counter = 5
```

## Plotting of train database

```python
X_train = df[(df['Fail Counter']==counter) & (df['Fail']==False)]
X_train = X_train.set_index('Cycle')
X_train = X_train[df.columns[3:9]]
X_train.plot(figsize=(800*pixel, 600*pixel))
plt.show()
```

## Plotting of test database

```python
X_test = df[(df['Fail Counter']==counter+1) & (df['Fail']==True)]
X_test = X_test.set_index('Cycle')
X_test = X_test[df.columns[3:9]]
X_test.plot(figsize=(800*pixel, 600*pixel))
plt.show()
```

## In order to detect anomaly, it is used an Autoencoder model, using a 3 layer neural network with 10-2-10 nodes.  Then, it is used softplus as activattion function, mean absolute error as loss function, and train the model using the Adam optimizer.

```python
seed(10)
set_seed(10)
act_func = 'softplus'
loss_func = 'mae'
opt_func = 'adam'

model=Sequential()

model.add(Dense(10,activation=act_func,
                kernel_initializer='glorot_uniform',
                kernel_regularizer=regularizers.l2(0.0),
                input_shape=(X_train.shape[1],)
               )
         )

model.add(Dense(2,activation=act_func,
                kernel_initializer='glorot_uniform'))

model.add(Dense(10,activation=act_func,
                kernel_initializer='glorot_uniform'))

model.add(Dense(X_train.shape[1],
                kernel_initializer='glorot_uniform'))

model.compile(loss=loss_func,optimizer=opt_func)

NUM_EPOCHS=400
BATCH_SIZE=5
```

## Also, it is used 400 epochs with a batch size of 5. Then, it is used 5% of the training data to validate each epoch.

```python
history=model.fit(np.array(X_train),np.array(X_train),
                  batch_size=BATCH_SIZE,
                  epochs=NUM_EPOCHS,
                  validation_split=0.05,
                  verbose = 1)
```

```python
plt.plot(history.history['loss'],
         'b',
         label='Training loss')
plt.plot(history.history['val_loss'],
         'r',
         label='Validation loss')

plt.legend(loc='upper right')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.show()
```

## Then, with the distribution plot of the training set's calculated loss, it is possible to identify a threshold to detect anomalies, which means finding a value above the "noise level", and the detected anomalies should have a statistical significancy and rely above the "noise background"

```python
X_pred = model.predict(np.array(X_train))
X_pred = pd.DataFrame(X_pred,
                      columns=X_train.columns)
X_pred.index = X_train.index

scored = pd.DataFrame(index=X_train.index)
scored['Loss_mae'] = np.mean(np.abs(X_pred-X_train), axis = 1)
plt.figure()
sns.distplot(scored['Loss_mae'],
             bins = 50,
             kde= True,
             color = 'blue')
```

## Analyzing the the above loss distribution, a threshold of 30 to detected anomaly fits like a good value. Then, it is possible to calculate the loss in the test set to check if it crosses the threshold

```python
Threshold = 30
```

```python
X_pred = model.predict(np.array(X_test))
X_pred = pd.DataFrame(X_pred,
                      columns=X_test.columns)
X_pred.index = X_test.index

scored = pd.DataFrame(index=X_test.index)
scored['Loss_mae'] = np.mean(np.abs(X_pred-X_test), axis = 1)
scored['Threshold'] = Threshold
scored['Anomaly'] = scored['Loss_mae'] > scored['Threshold']
scored.head()
```

## Then, it is possible to do the same for the train test

```python
X_pred_train = model.predict(np.array(X_train))
X_pred_train = pd.DataFrame(X_pred_train,
                            columns=X_train.columns)

X_pred_train.index = X_train.index
scored_train = pd.DataFrame(index=X_train.index)
scored_train['Loss_mae'] = np.mean(np.abs(X_pred_train-X_train),axis = 1)
scored_train['Threshold'] = Threshold
scored_train['Anomaly'] = scored_train['Loss_mae'] > scored_train['Threshold']

scored = pd.concat([scored_train, scored])
```

## Finally, the entire plotting

```python
scored.plot(figsize = (800*pixel,600*pixel), color = ['blue','red'])
plt.show()
```

## In order to check false positives, false negatives, true positives, and true negatives. It is build a confusion matrix

```python
Anomaly = scored.Anomaly
Fail = df[((df['Fail Counter']==counter+1) & (df['Fail']==True)) |
          ((df['Fail Counter']==counter) & (df['Fail']==False))].set_index('Cycle').Fail
```

```python
df_confusion = pd.crosstab(Fail, Anomaly)
df_confusion
```

## In order to measure a good balance between precision and recall, it is calculated the F1 score.

```python
TP = df_confusion[True][True]
FPs = df_confusion[True][False]
TN = df_confusion[False][False]
FN = df_confusion[False][True]

confusion_matrix_accuracy = (TP + TN)/(TP + TN + FP + FN)
confusion_matrix_precision = TP/(TP + FP)
confusion_matrix_recall = TP/(TP + FN)

f1_score = 2*confusion_matrix_precision*confusion_matrix_recall/(confusion_matrix_precision + confusion_matrix_recall)
print(f1_score)
```
