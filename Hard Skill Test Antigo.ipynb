# 1 - Calculate how many times the equipment has failed 

```python
num_fails, first, num_fails_vector = 0, True, []
for i in df.Fail:
    if i and first:
        num_fails = num_fails +1
        first = False
    if not i: first = True
    num_fails_vector.append(num_fails)
df['Fail Counter'] = pd.Series(num_fails_vector)
```

# 3 - Categorize equipment failures by their nature/root cause according to parameter readings (temperature, pressure, and others)



```python
df[df.columns[3:9]].boxplot(figsize=(800*pixel, 600*pixel))
plt.show()
```

```python
def check_outliers(df, df_Fail, k = 1.5):
    stats_parameters = df[df.columns[3:9]].describe()
    upper_dict = {}
    for column in stats_parameters.columns:
        IQR = stats_parameters[column]['75%']-stats_parameters[column]['25%']
        upper_dict[column] = stats_parameters[column]['75%'] + k*(IQR)

    stats_parameters = stats_parameters.append(pd.Series(data=upper_dict, name ='upper extreme'))
    display(stats_parameters)
    

    columns = df.columns[3:9]
    for column in columns:
        outlier_class = {}
        for i in df_Fail[column].index:
            if df_Fail[column][i] <= stats_parameters[column]['25%']: outlier_class[i] = 'Q1'
            if df_Fail[column][i] > stats_parameters[column]['25%'] and df_Fail[column][i] <= stats_parameters[column]['50%']: outlier_class[i] = 'Q2'
            if df_Fail[column][i] > stats_parameters[column]['50%'] and df_Fail[column][i] <= stats_parameters[column]['75%']: outlier_class[i] = 'Q3'
            if df_Fail[column][i] > stats_parameters[column]['75%'] and df_Fail[column][i] <= stats_parameters[column]['upper extreme']: outlier_class[i] = 'Q4'
            if df_Fail[column][i] > stats_parameters[column]['upper extreme']: outlier_class[i] = 'outlier'
        outlier_series = pd.Series(data=outlier_class, name =column+'_outlier')
        outlier_series.index = outlier_series.index.set_names('Cycle')

        df_Fail[column+'_outlier'] = outlier_series
        
    display(df_Fail) #pd.set_option("max_rows", None) pd.set_option('max_columns', None)
    
    df_Fail['Nature/Root'] = None
    for i in df_Fail.index:
        root = ''
        if df_Fail['Temperature_outlier'][i] == 'outlier': root = root+' Temperature'
        if df_Fail['Pressure_outlier'][i] == 'outlier': root = root+' Pressure'
        if df_Fail['VibrationX_outlier'][i] == 'outlier': root = root+' VibrationX'
        if df_Fail['VibrationY_outlier'][i] == 'outlier': root = root+' VibrationY'
        if df_Fail['VibrationZ_outlier'][i] == 'outlier': root = root+' VibrationZ'
        if df_Fail['Frequency_outlier'][i] == 'outlier': root = root+' Frequency'
        df_Fail['Nature/Root'][i] = root

    display(df_Fail['Nature/Root'].value_counts())
    
    #return df_Fail
```

```python
check_outliers(df, df_Fail)
```

```python
check_outliers(df, df_Fail, 1.0)
```

```python
check_outliers(df, df_Fail, 0.50)
```

```python
check_outliers(df, df_Fail, 0.40)
```
